```python
import kagglehub
path = kagglehub.dataset_download("shivamb/netflix-shows")
import pandas as pd
df = pd.read_csv('ipl.csv')
df.head()
```

//Data Preprocessing//

Changing the name of all team which has change there name in future


```python
name_map = {
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Kings XI Punjab': 'Punjab Kings'
}
cols = ['team1', 'team2', 'toss_winner', 'winner']

for col in cols:
    df[col] = df[col].replace(name_map)
```

```python
df['loser'] = df.apply(
    lambda row: row['team2'] if row['winner'] == row['team1'] else row['team1'],
    axis=1
)
df.drop(['player_of_match','toss_decision','toss_winner','match_date',
         'venue','match_type','result','super_over'],
        axis=1, inplace=True, errors='ignore')
df.head()
```

```python
df.nlargest(5, 'result_margin')
```

```python
df.drop(['team1_players','team2_players','city','match_number','result_margin','target_runs','target_overs'], axis=1, inplace=True, errors='ignore')
df.tail()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

df_clean = df[['team1', 'team2', 'season', 'winner']].dropna()

X = df_clean[['team1', 'team2', 'season']]
y = df_clean['winner']


X_train = X[X['season'] <= 16]
y_train = y[X['season'] <= 16]

X_test = X[X['season'] > 16]
y_test = y[X['season'] > 16]

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_train_enc = encoder.fit_transform(X_train)
X_test_enc = encoder.transform(X_test)

model = RandomForestClassifier(n_estimators=300, max_depth=15, random_state=42)
model.fit(X_train_enc, y_train)


y_pred = model.predict(X_test_enc)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))


results = X_test.copy()
results['Actual_Winner'] = y_test.values
results['Predicted_Winner'] = y_pred

print("\nSample Predictions:")
print(results)
```
